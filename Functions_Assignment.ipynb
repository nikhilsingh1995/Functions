2. Explain the concept of function arguments and parameters in Python.

Answer:

In Python, function parameters and arguments play a key role in making functions flexible and reusable.

 **Parameters** are the variable names listed in a function’s definition. They act as placeholders for the values that the function will operate on.

On the other hand, **arguments** are the actual values we pass to the function when we call it.

For example, in the function definition def greet(name):, name is a parameter. When we call greet("Nikhil"), the string "Nikhil" is the argument being passed to the name parameter.

Parameter = placeholder (like a variable in the function definition).

Argument = real value passed to that placeholder.

````python
# 'name' is a parameter
def greet(name):
    print("Hello,", name)

# 'Nikhil' is an argument
greet("Nikhil")
````

4. What is the purpose of the `return` statement in a Python function?

Answer:

The `return` statement in a Python function is used to **send back a value** from the function to the caller. It ends the function’s execution and provides the result that the function has computed. This is essential when we want to use the output of a function later in our code, such as storing it in a variable, printing it, or using it in another calculation.




**Key Purposes of `return`:**

1. **Pass result back to the caller** – Functions often perform a calculation or data processing, and `return` is how they give the result back.

2. **End the function execution** – Once a `return` is executed, the function exits immediately.

3. **Support reusability** – By returning values, you can call the function in different parts of the program and use the result flexibly.





**Example:**

```python
def add(a, b):
    return a + b  # Returns the sum

result = add(50, 50)
print("The sum is:", result)
```

**Output:**

```
The sum is: 100
```

Without a `return`, the function would just perform the operation but not give back the result to be reused.



**Optional Return:**

If no `return` is used, Python implicitly returns `None`:

```python
def do_nothing():
    pass

print(do_nothing())  # Output: None
```



To sum it up, the `return` statement makes functions **productive and reusable** by allowing them to pass data back to the rest of your program.


5. What are iterators in Python and how do they differ from iterables?

Answer:

In Python, **iterables** and **iterators** are fundamental concepts used in loops, comprehensions, and many data-processing tasks.



**Iterable:**

An **iterable** is any Python object that **can be looped over** (i.e., we can use it in a `for` loop). Examples include lists, tuples, strings, sets, and dictionaries. Internally, an iterable implements the `__iter__()` method, which returns an **iterator**.

Example of an Iterable:

```python
my_list = [1, 2, 3]
for item in my_list:
    print(item)
```

Here, `my_list` is an iterable.

---

**Iterator:**

An **iterator** is an object that represents a **stream of data**, returned one element at a time by calling the built-in `next()` function. It **remembers its state** and moves forward with each call to `next()`. Iterators implement two methods: `__iter__()` and `__next__()`.

You can get an iterator from any iterable by passing it to the `iter()` function.

Example of an Iterator:

```python
my_list = [1, 2, 3]
my_iter = iter(my_list)  # Convert iterable to iterator

print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3
```

After the last element, calling `next(my_iter)` again will raise a `StopIteration` exception.




6. Explain the concept of generators in Python and how they are defined

Answer:

A **generator** in Python is a special type of iterator that allows us to **generate values on the fly** rather than storing them all in memory at once.
It’s a way to write iterators more simply and efficiently, especially when working with large datasets or infinite sequences.

---

Key Concept:

* **Generators** produce values **lazily**, meaning one at a time and only when needed.
* Defined using a **function** with the `yield` keyword instead of `return`.

---

Generator Defination:

We define a generator just like a regular function but use `yield` instead of `return`.
Each time the generator function is called using `next()`, it resumes from where it left off after the last `yield`.

Example:

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

#### Output:

```
1
2
3
4
5
```

Using Generators Manually:

```python
gen = count_up_to(3)
print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
```

---

Summary:

Generators are perfect for creating **efficient, memory-friendly code** that processes sequences lazily.
They’re commonly used in reading large files, streaming data, or implementing custom iterators.




7. What are the advantages of using generators over regular functions?

Answer:

Generators offer several advantages over regular functions, especially when working with large datasets, streams, or operations that benefit from **lazy evaluation** (processing data one item at a time).
 Here are details:

---

**A. Memory Efficiency**

Generators **don’t store all values in memory**. Instead, they yield one value at a time, which is particularly useful when working with large or infinite sequences.

Example:

```python
def generate_numbers():
    for i in range(1000000):
        yield i
```

This uses almost no memory, unlike a list that stores all values.

---

**B. Lazy Evaluation**

Generators **produce values only when needed**, which can significantly reduce computation time and memory usage for large or complex operations.

We can start processing results immediately without waiting for the entire output to be generated.

---

 **C. Cleaner, Simpler Code**

Generators allows usto **write iterators without implementing `__iter__()` and `__next__()`** manually. This makes our codes more readable and maintainable.

Example:

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

Much simpler than writing a full iterator class!

---

**D. Infinite Sequences Support**

Generators are ideal for producing **infinite sequences** like Fibonacci numbers, sensor data, or live feeds, where we don’t know the total number of items in advance.

Example:

```python
def infinite_counter():
    i = 0
    while True:
        yield i
        i += 1
```

---

**E. Pipeline-Friendly (Composability)**

Generators can be **chained or composed** together to create powerful data-processing pipelines without holding intermediate results in memory.

```python
def square(numbers):
    for n in numbers:
        yield n * n

result = square(range(5))
```

---
 **F. Better Performance**

For I/O-bound or CPU-light tasks, generators can **increase performance** by reducing unnecessary memory allocation and function call overhead.

---


In short, **generators are ideal when you want performance, scalability, and elegant code** for data-driven applications.



8. What is a lambda function in Python and when is it typically used?

Answer:

A **lambda function** in Python is a small, anonymous function defined using the `lambda` keyword. It can have any number of arguments but only **one expression**, which is evaluated and returned automatically. Lambda functions are also known as **inline functions** or **throwaway functions** because they are often used for short, one-time operations.
Lambda functions are commonly used in situations where a small function is needed **temporarily**, especially as an **argument to higher-order functions** (functions that take other functions as input).

---

Basic Syntax:

```python
lambda arguments: expression
```

Example:

```python
square = lambda x: x * x
print(square(5))  # Output: 25
```

This is equivalent to:

```python
def square(x):
    return x * x
```

---

Common Use Cases:

A. **With `map()` – apply a function to every element in an iterable:**

```python
numbers = [1, 2, 3]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # [1, 4, 9]
```

B. **With `filter()` – filter elements based on a condition:**

```python
numbers = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # [2, 4]
```

C. **With `sorted()` – custom sort logic:**

```python
words = ["apple", "banana", "kiwi"]
sorted_words = sorted(words, key=lambda word: len(word))
print(sorted_words)  # ['kiwi', 'apple', 'banana']
```

D. **With `reduce()` – perform cumulative operations (from `functools`):**

```python
from functools import reduce
product = reduce(lambda x, y: x * y, [1, 2, 3, 4])
print(product)  # 24
```

---

Advantages of Lambda Functions:

* Short and concise for simple operations.
* Useful in functional-style programming.
* Ideal for one-liner functions.

---

Disadvantages:

* Cannot contain multiple expressions or statements.
* Less readable for complex logic.
* No name (anonymous), so harder to debug or reuse.

9. Explain the purpose and usage of the `map()` function in Python

Answer:

The `map()` function in Python is a built-in higher-order function used to **apply a function to every item in an iterable** (like a list, tuple, etc.).
It returns a map object (an iterator), which can be converted to a list, set, or tuple.

---
**Syntax:**

```python
map(function, iterable)
```

* **`function`**: A function (can be a named function or lambda) that will be applied to each item in the iterable.
* **`iterable`**: A sequence like a list, tuple, or string.

---

**Example:**

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]
```

---

**Using `map()` with Lambda Function:**

```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

---

**Multiple Iterables:**

If the function takes more than one argument, we can pass **multiple iterables** to `map()`.
The function will be applied using the items from all iterables in parallel.

```python
a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)
print(list(result))  # Output: [5, 7, 9]
```

---

In Short:

* `map()` helps transform elements of an iterable using a function.
* It’s **efficient**, **clean**, and supports **functional programming** principles.
* Especially useful when we want to apply the same operation to every item in a sequence.
* `map()` returns a **map object**, so you must use `list()`, `set()`, etc., to get a usable collection.
* If the iterable is exhausted, reusing the map object won’t work unless it's recreated.


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Answer:

`map()`, `filter()`, and `reduce()` are **functional programming tools** in Python that operate on iterables.
While they are similar in that they apply a function to elements of a sequence, their purposes and behaviors are quite different.

---

`map()` – *Transform each item*

* **Purpose:** Applies a function to **each item** in an iterable and returns an iterator of the results.
* **Used in:** Data transformation (e.g., doubling numbers, formatting strings)

Example:

```python
numbers = [1, 2, 3]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6]
```

---

`filter()` – *Select items*

* **Purpose:** Applies a function that returns **True or False** to each item and **filters** out the ones for which the function returns False.
* **Used in:** Filtering based on conditions (e.g., even numbers, non-empty strings)

Example:

```python
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # Output: [2, 4]
```

---

`reduce()` – *Combine all items into one*

* **Purpose:** Applies a function cumulatively to **reduce** the iterable to a **single value**.
* **Used in:** Summing, multiplying, combining values (e.g., factorial, total sum)
* **Import:** Comes from the `functools` module

Example:

```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10
```

---

In Short:

* We use **`map()`** to *transform* data,
* We use **`filter()`** to *select* data,
* We use **`reduce()`** to *aggregate* data into a single result.

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
# Q4 Answer:

def is_prime(n):
    if n < 2 or n > 200:
        return False  # Only check for numbers in the range 2 to 200
    for i in range(2, int(n**0.5) + 1):  # Only check up to the square root of n
        if n % i == 0:
            return False
    return True

# Usage:
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")

#Numbers less than 2 are not prime.

#The function checks divisibility up to the square root of the number for efficiency.

#It returns True if no divisors are found, False otherwise.
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

````python
# Q5 Answer:

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)

#__init__: Initializes the iterator with the max number of terms.

#__iter__: Returns the iterator object (required for iteration).

#__next__: Generates the next Fibonacci number, and raises StopIteration after the limit.
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
# Q6 Answer:

def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

# Usage:
for value in powers_of_two(5):
    print(value)

#The powers_of_two function uses the yield keyword to generate values one at a time.

#Since it's a generator, it is memory-efficient and suitable for large sequences.
````

7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
# Q7 Answer:

# Step 1: Create the sample.txt file

def create_sample_file():
    content = """Line 1: Hello, World!
Line 2: This is a test file.
Line 3: Each line will be read using a generator.
Line 4: Python is awesome!
Line 5: End of file."""
    with open("sample.txt", "w") as file:
        file.write(content)

# Step 2: Generator function to read lines

def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

# Step 3: Run the process

create_sample_file()  # Making sure the file exists

print("Reading from sample.txt:\n")
for line in read_file_lines("sample.txt"):
    print(line)

#open(filename, 'r') opens the file in read mode.

#The for line in file loop reads one line at a time — this is memory-efficient.

#yield pauses the function and returns the line; it resumes from the same point on the next call.
````
